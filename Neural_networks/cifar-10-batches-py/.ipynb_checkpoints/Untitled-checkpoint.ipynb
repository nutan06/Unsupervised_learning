```python
from sklearn.neural_network import MLPClassifier

X_train = data[:300, :]
y_train = labels[:300]

X_test = data[300:400, :]
y_test = labels[300:400]

# Establish and fit the model, with a single, 3072 perceptron layer.
mlp = MLPClassifier(hidden_layer_sizes=(3072,), activation='logistic')
mlp.fit(X_train, y_train)

y_pred = mlp.predict(X_test)
```

```python
rgb_img = np.reshape(data[9],(32, 32, 3), order='F')

plt.plot()
plt.title('Figure 1 : RGB Image')
plt.imshow(rgb_img, interpolation='nearest');
```

```python
# Hyper-parameters. These were set by cross-validation,
# using a GridSearchCV. Here we are not performing cross-validation to
# save time.
logistic = linear_model.LogisticRegression()
rbm = BernoulliRBM(random_state=0)
classifier = Pipeline(steps=[('rbm', rbm), ('logistic', logistic)])
rbm.learning_rate = 0.06
rbm.n_iter = 20
# More components tend to give better prediction performance, but larger
# fitting time
rbm.n_components = 100
logistic.C = 6000.0

# Training RBM-Logistic Pipeline
classifier.fit(X_train, y_train)

# Training Logistic regression
logistic_classifier = linear_model.LogisticRegression(C=100.0)
logistic_classifier.fit(X_train, y_train)
```

```python
print()
print("Logistic regression using RBM features:\n%s\n" % (
    metrics.classification_report(
        y_test,
        classifier.predict(X_test))))

print("Logistic regression using raw pixel features:\n%s\n" % (
    metrics.classification_report(
        y_test,
        logistic_classifier.predict(X_test))))
```

```python
plt.figure(figsize=(15, 15))
for i in range (0,100):
    comp = rbm.components_[i]
    plt.subplot(10, 10, i + 1)
   
    plt.imshow(comp.reshape((32, 32,3)), 
               cmap=plt.cm.gray_r,
               interpolation='nearest')
    plt.xticks(())
    plt.yticks(())
plt.suptitle('100 components extracted by RBM', fontsize=16)


plt.show();
```
